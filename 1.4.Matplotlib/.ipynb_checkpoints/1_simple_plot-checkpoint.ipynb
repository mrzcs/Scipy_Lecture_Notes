```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

## 2.1 plotting with default settings

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

## 2.2 instantiating defaults

```python
# create a figure of size 8 x 6 inches, 80 dpi
plt.figure(figsize=(10, 6), dpi=80)

# create a new subplot from a gird of 1 x 1
plt.subplot(1, 1, 1)

# plot cos with a blue line of width 1 px
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")

# plot sin with a green line of width 1 px
plt.plot(X, S, color='green', linewidth=1.0, linestyle='-')

# set x limits
plt.xlim(-4.0, 4.0)

# set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

# set y limits
plt.ylim(-1.0, 1.0)

# set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# save figure using 72 dpi
#plt.savefig("name.png", dpi=72)

plt.show()
```

## 2.3 changing colors and line widths

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.subplot(1, 1, 1)
# set x limits
plt.xlim(-4.0, 4.0)
# set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# set y limits
plt.ylim(-1.0, 1.0)
# set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# changing colors and line widths
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")
plt.show()
```

## 2.4 setting limits

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.subplot(1, 1, 1)
# set x limits
#plt.xlim(-4.0, 4.0)
# set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# set y limits
#plt.ylim(-1.0, 1.0)
# set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# changing colors and line widths
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")

# setting limit
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.show()
```

## 2.5 setting ticks

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.subplot(1, 1, 1)
# set x limits
#plt.xlim(-4.0, 4.0)
# set x ticks
#plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# set y limits
#plt.ylim(-1.0, 1.0)
# set y ticks
#plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# changing colors and line widths
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")

# setting limit
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

#setting ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1],
          [r'$-1$', r'$0$', r'$+1$'])
plt.show()
```

## 2.7 moving spines

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.subplot(1, 1, 1)
# set x limits
#plt.xlim(-4.0, 4.0)
# set x ticks
#plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# set y limits
#plt.ylim(-1.0, 1.0)
# set y ticks
#plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# changing colors and line widths
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")

# setting limit
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

#setting ticks
plt.xticks([-np.pi, np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 1],
          [r'$-1$', r'$0$', r'$+1$'])


# moving spines
ax = plt.gca() # get current axis

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0)) # move x asix to coordinate 0 in data space coordinates.
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0)) # move y asix to coordinate 0 in data space coordinates.


plt.show()
```

## 2.8 adding a legend

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.subplot(1, 1, 1)
# set x limits
#plt.xlim(-4.0, 4.0)
# set x ticks
#plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# set y limits
#plt.ylim(-1.0, 1.0)
# set y ticks
#plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# changing colors and line widths
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="consine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

# adding a legend
plt.legend(loc='upper left')

# setting limit
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

#setting ticks
plt.xticks([-np.pi, np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 1],
          [r'$-1$', r'$0$', r'$+1$'])

# moving spines
ax = plt.gca() # get current axis

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0)) # move x asix to coordinate 0 in data space coordinates.
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0)) # move y asix to coordinate 0 in data space coordinates.


plt.show()
```

## 2.9 annotate some points

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.subplot(1, 1, 1)
# set x limits
#plt.xlim(-4.0, 4.0)
# set x ticks
#plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# set y limits
#plt.ylim(-1.0, 1.0)
# set y ticks
#plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# changing colors and line widths
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="consine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

# adding a legend
plt.legend(loc='upper left')

# setting limit
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

#setting ticks
plt.xticks([-np.pi, np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 1],
          [r'$-1$', r'$0$', r'$+1$'])

# moving spines
ax = plt.gca() # get current axis

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0)) # move x asix to coordinate 0 in data space coordinates.
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0)) # move y asix to coordinate 0 in data space coordinates.


# annotate some points
t = 2 * np.pi/3
plt.plot([t, t], [0, np.cos(t)], color="blue", linewidth=2.5, 
        linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color="blue")

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t, np.cos(t)), xycoords='data',
            xytext=(-90, -50), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))

plt.plot([t, t], [0, np.sin(t)], color="red", linewidth=2.5, 
        linestyle="--")
plt.scatter([t, ], [np.sin(t), ], 50, color="blue")

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))


plt.show()
```
